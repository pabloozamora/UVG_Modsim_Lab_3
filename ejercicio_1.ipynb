```python
import numpy as np

def runge_kutta_4(f, y0, t0, tf, h):
    """
    Implementa el método de Runge-Kutta de cuarto orden (RK4) para resolver una EDO.

    :param f: Función que describe la EDO (y' = f(t, y))
    :param y0: Valor inicial de y
    :param t0: Tiempo inicial
    :param tf: Tiempo final
    :param h: Paso de integración
    :return: Valores de t y y calculados en cada paso
    """
    n_steps = int((tf - t0) / h)
    t_values = np.linspace(t0, tf, n_steps + 1)
    y_values = np.zeros(n_steps + 1)
    
    y_values[0] = y0
    t = t0
    y = y0

    for i in range(1, n_steps + 1):
        k1 = h * f(t, y)
        k2 = h * f(t + h/2, y + k1/2)
        k3 = h * f(t + h/2, y + k2/2)
        k4 = h * f(t + h, y + k3)
        
        y += (k1 + 2*k2 + 2*k3 + k4) / 6
        t += h
        y_values[i] = y

    return t_values, y_values
```

```python
# Ejemplo

def f(t, y):
    return -2 * y + t

# Parámetros iniciales
y0 = 0
t0 = 0
tf = 5
h = 0.1

t_values, y_values = runge_kutta_4(f, y0, t0, tf, h)

for t, y in zip(t_values, y_values):
    print(f"t = {t:.2f}, y = {y:.5f}")
```

```python
def runge_kutta_4_system(f, y0, t0, tf, h):
    """
    Método de Runge-Kutta de cuarto orden para un sistema de EDOs.
    
    :param f: Función que describe el sistema de EDOs.
    :param y0: Valores iniciales para y (vector).
    :param t0: Tiempo inicial.
    :param tf: Tiempo final.
    :param h: Paso de integración.
    :return: Valores de t y y calculados en cada paso.
    """
    n_steps = int((tf - t0) / h)
    t_values = np.linspace(t0, tf, n_steps + 1)
    y_values = np.zeros((n_steps + 1, len(y0)))
    
    y_values[0, :] = y0
    t = t0
    y = np.array(y0, dtype=float)

    for i in range(1, n_steps + 1):
        k1 = h * f(t, y)
        k2 = h * f(t + h/2, y + k1/2)
        k3 = h * f(t + h/2, y + k2/2)
        k4 = h * f(t + h, y + k3)
        
        y += (k1 + 2*k2 + 2*k3 + k4) / 6
        t += h
        y_values[i, :] = y

    return t_values, y_values
```

```python
# Ejemplo

def system_of_eqs(t, y):
    """
    Define el sistema de ecuaciones diferenciales.
    y[0] = y1, y[1] = y2
    """
    dydt = np.zeros(2)
    dydt[0] = y[1]  # Ejemplo: dy1/dt = y2
    dydt[1] = -y[0] # Ejemplo: dy2/dt = -y1
    return dydt
```

```python
import matplotlib.pyplot as plt

# Parámetros iniciales
y0 = [1, 0]  # y1(0) = 1, y2(0) = 0
t0 = 0
tf = 10
h = 0.1

# Resolver el sistema de EDOs usando RK4
t_values, y_values = runge_kutta_4_system(system_of_eqs, y0, t0, tf, h)

# Graficar los resultados
plt.plot(t_values, y_values[:, 0], label="y1(t)")
plt.plot(t_values, y_values[:, 1], label="y2(t)")
plt.xlabel("Tiempo (t)")
plt.ylabel("Valores de y1 y y2")
plt.title("Sistema de EDOs resuelto con RK4")
plt.legend()
plt.grid(True)
plt.show()
```
